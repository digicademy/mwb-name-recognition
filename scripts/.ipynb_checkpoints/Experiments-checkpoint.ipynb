```python
import os
from glob import glob

for file in glob("../data/tsv/*.tsv"):
    print(os.path.splitext(file)[0])
```

```python
df = pd.DataFrame()
for file in glob("*.tsv"):
    temp_df = pd.read_csv(file, sep="\t")
    temp_df["source"] = file.replace(".tsv", "")
    df = pd.concat([df, temp_df])
df = df.dropna()
df
```

```python
df["label"].value_counts()
```

```python
uppercase_df = pd.DataFrame(df.loc[df["text"].str.contains("^[A-Z]")][["source", "text", "label"]])
uppercase_df
```
